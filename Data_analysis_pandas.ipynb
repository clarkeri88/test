```python
import pandas as pd

travel_times_df = pd.read_csv('travel-times.csv')
```

```python
#print shape
print(travel_times_df.shape)
```

```python
#print first 15 rows

print(travel_times_df.head(15))
```

```python
#Get info for the features (columns) with missing values
missing_values_num = travel_times_df.isnull().sum()
features_with_missing_values = missing_values_num[missing_values_num > 0]

print("Features with missing values and their counts:")
print(features_with_missing_values)

print("\nPercentage of missing values per feature:")
print(features_with_missing_values / len(travel_times_df) * 100)
```

```python
#Drop duplicate values by keeponig only last instance
travel_times_df.drop_duplicates(keep='last', inplace=True)
```

```python
#Total number of missing values
total_missing_values = travel_times_df.isnull().sum().sum()
print(f"Total number of missing values in the DataFrame: {total_missing_values}")
```
