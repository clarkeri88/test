## **Numerical Calculator (if-elif-else statements):**

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == "+":
  result = num1 + num2
elif operation == "-":
  result = num1 - num2
elif operation == "*":
  result = num1 * num2
elif operation == "/":
  result = num1 / num2
elif operation == "//":
  result = num1 // num2
elif operation == "%":
  result = num1 % num2
elif operation == "**":
  result = num1 ** num2
else:
  print("Invalid operation")
  result = None

print(f"result: {result}")
```

```python
def reverse(s):
  if len(s) < 2:
    return s
  else:
    return s[-1] + reverse(s[1:-1]) + s[0]

reverse('12345')

```

# **ROI CALC**

```python
def roi(cost, revenue):
  # Ensure inputs are numeric types
  assert isinstance(cost, (int, float)), "Cost must be an integer or a float."
  assert isinstance(revenue, (int, float)), "Revenue must be an integer or a float."

  # Ensure cost is not zero to avoid division by zero
  if cost == 0:
    raise ValueError("Cost cannot be zero for ROI calculation.")
  # Ensure cost is positive for a standard ROI interpretation
  if cost < 0:
    raise ValueError("Cost cannot be negative for standard ROI calculation.")

  ROI = (revenue - cost) / cost
  # Returns as a decimal (float)
  return ROI
```

# **Quality Control Manager (exceptions)??**

```python
def getValues(n):
  """Prompts the user for n values and returns a list of valid (non-negative, numeric) inputs."""
  values = []
  for i in range(n):
    while True:
      try:
        num_str = input(f"Enter value {i + 1}: ")
        num = float(num_str)
        if num < 0:
          raise ValueError("Value cannot be negative.")
        values.append(num)
        break
      except ValueError as e:
        print(f"Invalid input: {e}. Please enter a non-negative number.")
  return values

def calcAverage(alist):
  """Calculates the average of a list, excluding the first three values."""
  if len(alist) <= 3:
    print("List must contain more than 3 values to exclude the first three.")
    return 0.0 # Or raise an error, depending on desired behavior

  # Exclude the first three values
  trimmed_list = alist[3:]

  if not trimmed_list:
    return 0.0 # Avoid division by zero if all values were excluded

  return sum(trimmed_list) / len(trimmed_list)

# Main program execution
if __name__ == "__main__":
  num_inputs = 6 # Example: Ask for 6 values, so we have at least 3 to average after exclusion

  print(f"Please enter {num_inputs} numeric values for quality control.")
  weights = getValues(num_inputs)

  print(f"Original values: {weights}")

  if len(weights) > 3:
    average_weight = calcAverage(weights)
    print(f"Average weight (excluding first three): {average_weight:.2f}")
  else:
    print("Not enough valid values entered to calculate an average after exclusion.")
```

# **Student Class**

```python
class Student:
  def __init__(self,name,number):
    self.__name = name
    self.number = number
    self.__courses = []

  def get_courses(self):
    return self.__courses

  def get_name(self):
    return self.__name

  def set_name(self,newName):
    self.__name =newName

  def __len__(self):
    return len(self._courses)

  def enroll(self,course):
    if course in self.__courses:
      print(f"Student {self.__name} is already enrolled in {course}")
    else:
      self.__courses.append(course)
```

# **Class Point**

```python
class point:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def distance(self,other):
    dx = (self.x - other.x)**2
    dy = (self.y - other.y)**2
    dist = (dx+dy)**.05
    return dist
class Rectangles:
  def __init__(self,x1,y1,x2,y2, x3,y3,x4,y4):
    self.p1 = point(x1,y1)
    self.p2 = point(x2,y2)
    self.p3 = point(x3,y3)
    self.p4 = point(x4,y4)
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)
  def area(self):
    return self.width * self.height
  def perimeter(self):
    return 2*self.width + 2*self.height
```

# **Class employee -manageer  WEEK 14**

```python
class Employee:
  def __init__(self, name, hourly_wage, hours_worked,):
    self.name = name
    self.hourly_wage = hourly_wage
    self.hours_worked = hours_worked

  def calculate_pay(self):
    return self.hourly_wage * self.hours_worked

  def __str__(self):
    return f"Emoloyee name: {self.name} \nHourly charge: {self.hourly_wage}$"

  def get_wage(self):
    return self.hourly_wage

  def set_wage(self,new_wage):
    assert isinstance(new_wage,(int,float))
    assert new_wage >0
    if new_wage <=0:
      raise ValueError("Invalid wage")
    self.hourly_wage = new_wage
```

```python
#part 3

class Manager(Employee):
  def __init__ (self,name, hourly_wage, hours_worked, bonus):
    super().__init__(name, hourly_wage, hours_worked)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage()


employee = Employee("Bob",50,10)
manager = Manager("Bob",50,10,0.2)

print(employee)
print(manager)
print()
print (f"employee pay: {employee.name} weekly salary: {employee.calculate_pay()}$")
```
