### **Income Statement using Simple Income Statement Calculation:**

```python
Sales = float(input("Enter Sales($)"))
cogs = Sales*.40
Profit = Sales-cogs
print (f"Sales: ${Sales}")
print(f"cogs: ${cogs}")
print(f"Profit: ${Profit}")

tax_rate= int(input("Tax Rate? (%)"))
profit_after_tax = Profit-(Profit*(tax_rate/100))
print(f"Profit After Tax: ${profit_after_tax}")
```

## **Numerical Calculator (if-elif-else statements):**

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == "+":
  result = num1 + num2
elif operation == "-":
  result = num1 - num2
elif operation == "*":
  result = num1 * num2
elif operation == "/":
  result = num1 / num2
elif operation == "//":
  result = num1 // num2
elif operation == "%":
  result = num1 % num2
elif operation == "**":
  result = num1 ** num2
else:
  print("Invalid operation")
  result = None

print(f"result: {result}")
```

# **Rock-Paper-Scissors**

```python
import random

play = input("do you want to play? (y/n): ").lower()

while play == "y":
  p_move = input("choose between rock paper and scissors").lower()
  c_move = random.choice(["rock","paper","scissors"])

  print(f"computer chose {c_move}")
  print(f"player chose {p_move}")

  if p_move == c_move:
    print("tie")
  elif p_move == "rock" and c_move == "scissors":
    print("player wins")
  elif p_move == "paper" and c_move == "rock":
    print("player wins")
  elif p_move == "scissors" and c_move == "paper":
    print("player wins")
  else:
    print("computer wins")
  play = input("do you want to play? (y/n): ").lower()
```

# **Class employee -manageer  WEEK 14**

```python
class Employee:
  def __init__(self, name, hourly_wage, hours_worked,):
    self.name = name
    self.hourly_wage = hourly_wage
    self.hours_worked = hours_worked

  def calculate_pay(self):
    return self.hourly_wage * self.hours_worked

  def __str__(self):
    return f"Emoloyee name: {self.name} \nHourly charge: {self.hourly_wage}$"

  def get_wage(self):
    return self.hourly_wage

  def set_wage(self,new_wage):
    assert isinstance(new_wage,(int,float))
    assert new_wage >0
    if new_wage <=0:
      raise ValueError("Invalid wage")
    self.hourly_wage = new_wage
```

```python
#part 3

class Manager(Employee):
  def __init__ (self,name, hourly_wage, hours_worked, bonus):
    super().__init__(name, hourly_wage, hours_worked)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage()


employee = Employee("Bob",50,10)
manager = Manager("Bob",50,10,0.2)

print(employee)
print(manager)
print()
print (f"employee pay: {employee.name} weekly salary: {employee.calculate_pay()}$")
```
