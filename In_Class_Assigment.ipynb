```python
Sales = float(input("Enter Sales($)"))
cogs = Sales*.40
Profit = Sales-cogs
print (f"Sales: ${Sales}")
print(f"cogs: ${cogs}")
print(f"Profit: ${Profit}")

tax_rate= int(input("Tax Rate? (%)"))
profit_after_tax = Profit-(Profit*(tax_rate/100))
print(f"Profit After Tax: ${profit_after_tax}")
```
